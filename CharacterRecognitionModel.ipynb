```python
from tensorflow.keras.models import load_model
from google.colab import drive
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/drive',force_remount=True)
model = load_model('/content/drive/MyDrive/TASKS/CharacterRecognitionModel.h5')
```

```python
def Prediction(image):
    predictions = model.predict(image)
    return np.argmax(predictions)
```

```python
def Load_Image(image_path):
    img = Image.open(image_path).convert('L')
    img = img.resize((28, 28))
    img_array = np.array(img)
    img_array = img_array.astype('float32') / 255
    img_array = img_array.reshape(1, 28, 28, 1)
    return img_array
```

```python
def Show_Image(image_path):
    img = Image.open(image_path)
    plt.figure(figsize=(6, 3))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(img, cmap='gray')
    img_array = Load_Image(image_path).reshape(28, 28)
    plt.subplot(1, 2, 2)
    plt.title("Preprocessed Image")
    plt.imshow(img_array, cmap='gray')
    plt.show()
```

```python
image_path = '/content/drive/MyDrive/TASKS/IMAGE2.png'
loaded_image = Load_Image(image_path)
```

```python
predicted_label = Prediction(loaded_image)
print(f'Predicted label: {predicted_label}')
print("\n \n")
Show_Image(image_path)
```
